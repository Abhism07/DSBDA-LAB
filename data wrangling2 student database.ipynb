```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    'Name': ['John', 'Emma', 'Liam', 'Olivia', 'William', 'Sophia', 'Michael'],
    'Age': [18, 19, 17, 18, 19, 20, 17],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'Maths': [85, 92, 78, 90, 88, np.nan, 95],
    'Science': [90, 88, 92, 85, 80, 120, 92],
    'English': [82, 85, 88, 90, 87, 78, np.nan]
}
```

```python
df=pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
numeric_values = df[['Age', 'Maths', 'Science', 'English']]
```

```python
for values in numeric_values.columns:
    sns.boxplot(x=values,data=numeric_values)
    plt.title(values)
    plt.show()
```
